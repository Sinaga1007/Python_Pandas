```python
#Terlebih dahulu mengimport modules

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Mempersiapkan Data Frame

n_rows = 365 * 24       #24 merepresentasikan jam, 365 merepresentasikan hari
n_cols = 2
cols = ['col1','col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), #Bernilai integer yang dibuat secara acak
                  columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H') #Deret Waktu
df
```

```python
#Resampling data dengan interval monthly

df.resample('M')['col1'].sum().to_frame() #Untuk mengelompokkan per-bulan
```

```python
#Resampling data dengan interval daily

df.resample('D')['col1'].sum().to_frame() #Untuk mengelompokkan per-hari
```
