```python
#Terlebih dahulu mengimport modules

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Mempersiapkan Data Frame

n_rows = 7
n_cols = 6
cols = tuple('ABCDEF')

df = pd.DataFrame(np.random.randint(1, 7, size=(n_rows, n_cols)),
                  columns=cols)
df
```

```python
#Mengacak urutan baris pada Data Frame

df.sample(frac=1.0, random_state=1) #frac=1.0 artinya kita akan melakukan sampling dengan proporsi yang penuh
```
