```python
#Terlebih dahulu mengimport modules

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Mempersiapkan data frame

n_rows = 15
n_cols = 6
cols = tuple('ABCDEF')

df = pd.DataFrame(np.random.randint(1, 15, size=(n_rows, n_cols)),
                  columns=cols) #Untuk membuat bilangan random
df
```

```python
#Seleksi baris dengan between()

df[df['A'].between(2, 5)] #Fungsi between sifatnya inclusive artinya nilai 2 dan 5 akan disertakan  
```

```python
df[df['A'].between(2, 5, inclusive=False)] #Nilai 2 dan 5 sebagai pembatas saja
```
