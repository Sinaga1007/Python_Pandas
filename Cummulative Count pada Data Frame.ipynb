```python
#Terlebih dahulu mengimport modules

import pandas as pd

print(pd.__version__)
```

```python
#Mempersiapkan Data Frame

d = {'Pedagang':['Baji', 'Tono', 'Wani', 'Baji', 'Stev', 'Tono', 'Wani', 'Baji'],
     'Barang':['Monitor', 'Flashdisk', 'Keyboard', 'Mousepad', 'Keyboard', 'Monitor', 'Laptop', 'Monitor']}

df = pd.DataFrame(d)
df
```

```python
#Mengenal Cummulative Count dengan cumcount()

df['count_tiap_pedagang'] = df.groupby('Pedagang').cumcount() + 1  #Cumulative count jika tidak ditambah 1 maka akan mulai dari 0
df
```

```python
df['count_tiap_barang'] = df.groupby('Barang').cumcount() + 1
df
```

```python
df['count_pasangan_kolom'] = df.groupby(['Pedagang', 'Barang']).cumcount() + 1
df
```
