```python
#Terlebih dahulu kita mengimport modules
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Membuat data frame
n_rows = 6
n_cols = 6
cols = tuple('ABCDEF') #row dan col akan diberi label ABCDEF

#Membangun sebuah objek data frame
df = pd.DataFrame(np.random.randint(1,10, size=(n_rows, n_cols)), #Random randint ini bilangan bulat acak
                 columns=cols)
df
```

```python
#Membalik Urutan Kolom yang diatas, nah caranya seperti dibawah
df.loc[:, ::-1] #Bagian pertama bertugas untuk mengatur baris, bagian kedua bertugas untuk mengatur kolom

#-1 digunakan untuk pembalikan urutan
```

```python
#Membalik urutan baris
df.loc[::-1]
```

```python
#Membalik Urutan Baris dan Melakukan Penyesuaian Ulang Index
df.loc[::-1].reset_index(drop=True) #Berfungsi mengubah Index 0 jadi paling atas dan index ke 5 jadi paling bawah
```
