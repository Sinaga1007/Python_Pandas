```python
#Terlebih dahulu mengimport modules

import pandas as pd

print(pd.__version__)
```

```python
#Mempersiapkan Data Frame

df = pd.DataFrame({'Jenis_Kelamin':['Pria', 'Wanita', 'Lelaki', 'Lelaki', 'Perempuan'],
                   'Usia':[20, 22, 25, 23, 21],
                   'Shift':['Pagi', 'Siang', 'Malam', 'Siang', 'Pagi']})
df
```

```python
#Mapping Pada Data Framde dengan applymap()

df = df.applymap(lambda x: x.lower() if type(x) == str else x)  #Apabila nilai-nya string akan kita ubah menjadi lower case, ttpi jika bukan string akan dikembalikan apa adanya
df
```

```python
#Applymap() dengan dictionary

mapping = {'pria':'L',
           'lelaki':'L',
           'wanita':'P',
           'perempuan':'P',
           'pagi':1,
           'siang':2,
           'malam':3}

df.applymap(mapping.get)
```

```python
df[['Jenis_Kelamin', 'Shift']] = df[['Jenis_Kelamin', 'Shift']].applymap(mapping.get)

df
```
