```python
#Terlebih dahulu mengimport modules

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Mempersiapkan data frame

n_rows = 6
n_cols = 6
cols = tuple('ABCDEF') #Setiap kolom akan diberi label

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), #Untuk mengacak nilai integer
                  columns=cols)
df.head()
```

```python
#Bagian pertama

df1 = df.copy(deep=True)
df1 = df1.drop([1,4])
df1
```

```python
#Bagian kedua

df2 = df.copy(deep=True)
df2 = df2.drop([0,3])
df2
```

```python
#Menggabungkan Dua Data Frame

df_inner = pd.merge(df1, df2, how='inner') #Akan menggabungkan dua buah data frame dimana data frame yang baru akan berisi baris yang terdapat pada kedua data frame sumbernya
df_inner
```

```python
df_outer = pd.merge(df1, df2, how='outer') #Akan menggabungkan dua buah data frame dimana nilai duplicatenya akan dihilangkan
df_outer
```
