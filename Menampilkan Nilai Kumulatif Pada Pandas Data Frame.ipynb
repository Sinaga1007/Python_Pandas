```python
#Terlebih dahulu mengimport modules

import pandas as pd

print(pd.__version__)
```

```python
#Mempersiapkan Data Frame

d = {'Pemain':['Lionel', 'Jhoni', 'Mathew', 'Budi', 'Budi', 'Mathew', 'Asep', 'Jhoni'],
     'Goal':[3, 2, 4, 2, 2, 4, 4, 3]}

df = pd.DataFrame(d)
df
```

```python
#Menampilkan nilai Kumulatif (Cummulative)

df['Goal'].cumsum().to_frame()  #Cumsum berfungsi untuk mendata jumlah goal secara kumulatif 
```

```python
df['Jumlah_Goal_Kumulatif'] = df['Goal'].cumsum()
df
```

```python
df['Jumlah_Goal_Kumulatif_Tiap_Pemain'] = df.groupby('Pemain')['Goal'].cumsum()
df
```

```python
df['cummax'] = df['Goal'].cummax()
df
```

```python
df['cummin'] = df['Goal'].cummin()
df
```

```python
df['cumprod'] = df['Goal'].cumprod()
df
```
