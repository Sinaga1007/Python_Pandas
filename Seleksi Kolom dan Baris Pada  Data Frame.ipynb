```python
#Terlebih dahulu mengimport Modules

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Mempersiapkan sebuah data Frame

n_rows = 12
n_cols = 6
cols = tuple('ABCDEF')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)
df
```

```python
#Seleksi kolom dan baris menggunakan loc

df.loc[[0,3,4], ['B','E']] #Bagian pertama loc digunakan untuk baris, kedua untuk mengatur kolom
```

```python
#Menyeleksi baris dengan kondisi

df.loc[df['B']>10, ['B', 'D', 'E']] #Barisnya harus diatas 10
```

```python
#Slicing Data Frame dengan Loc

df.loc[0:4, 'B':'D'] #Melakukan slicing dari baris 0-4
```
