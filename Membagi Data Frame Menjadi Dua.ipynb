```python
#Terlebih dahulu mengimport Modules-nya

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Persiapan Data Frame

n_rows = 12 #Baris
n_cols = 6 #Kolom
cols = tuple ('ABCDEF')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),  #Berfungsi Untuk Mengacak Data
                  columns=cols)
df
```

```python
#Membagi Data Frame Menjadi Dua Secara Acak Berdasarkan Proporsi Tertentu

df.shape #Cek shape dari data frame kita
```

```python
#Membagi dua data frame

proporsi = 0.7 #70% untuk data frame pertama dan 30% untuk data frame kedua
df_1 = df.sample(frac=proporsi) #df.sample ini merupakan method yang akan menganmbil baris secara acak dari baris-baris yang dimiliki oleh data frame
df_2 = df.drop(df_1.index) #Akan menghapus data-data frame yang diadopsi dari df_1

print(f'df_1 shape: {df_1.shape}')
print(f'df_2 shape: {df_2.shape}')
```
