```python
#Terlebih dahulu mengimport modules

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Mempersiapkan Data Frame

n_rows = 45
n_cols = 6
cols = tuple('ABCDEF')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),  #Berupa nilai random
                  columns=cols)
df.head()
```
