```python
#Terlebih dahulu mengimport modules

import pandas as pd

print(pd.__version__)
```

```python
d = {'Pedagang':['Anto', 'Toni', 'Ani', 'Anto', 'Boby', 'Toni', 'Ani', 'Anto'],
     'Barang':['Keyboard', 'Keyboard', 'Mouse', 'Monitor', 'Mouse', 'Keyboard', 'Laptop', 'Keyboard']}

df = pd.DataFrame(d)
df
```

```python
#Membuat kolom baru dengan for loop dan f-string

cols = ['Pedagang', 'Barang'] #Menyiapkan variabel yang menampung list

for col in cols:
    df[f'count_tiap_{col}'] = df.groupby(col).cumcount() + 1    #Untuk melakukan groupping
    
df
```
