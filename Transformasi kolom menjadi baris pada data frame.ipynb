```python
#Terlebih dahulu mengimport modules

import pandas as pd

print(pd.__version__)
```

```python
#Mempersiapkan data frame

d = {
'Kode_Area':[134, 500, 799, 543],
'Pabrik':[3, 5, 9, 1],
'Gudang':[2, 7, 6, 2],
'Toko':[57, 35, 22, 29]
}

df = pd.DataFrame(d)
df
```

```python
#Transformasi kolom menjadi baris

df = df.melt(id_vars='Kode_Area',         #id_vars ini identik sebagai kolom unik dari suatu data frame
             var_name='jenis_bangunan',   
             value_name='jumlah')         
df
```
