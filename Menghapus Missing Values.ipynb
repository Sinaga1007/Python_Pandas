```python
#Terlebih dahulu mengimport pandasnya
import pandas as pd

print(pd.__version__)
```

```python
#Membuat data frame yang missing value
df = pd.util.testing.makeMissingDataframe().reset_index()
df.head()
```

```python
#Mengubah nama kolom indexnya
df = df.rename( columns={'index':'Z'})
df.head()
```

```python
#Membuat copy dari data diatas dan disimpan di df_backup
df_backup = df.copy( deep=True)
```

```python
#Menghapus (drop) setiap kolom yang mengandung missing values
df = df.dropna(axis='columns') #Parameter colum ini berarti kita akan mendrop columnya
df.head()
```

```python
#Menghapus (drop) setiap baris yang mengandung missing values
df = df_backup.copy(deep=True) #df dengan kata lain menampung data original yang pertama kali kita buat sebelumnya
df = df.dropna(axis='rows') #Berfungsi untuk menghapus rows yang mengandung missing values
df.head()
```

```python
#Persentase missing values untuk tiap kolom
df = df_backup.copy(deep=True)
df.isna().mean() #df.isna berfungsi untuk mengumpulkan semua missing values untuk setiap kolom pada data frame tersebut
                #mean berfungsi untuk mengkalkulasi rata-rata/persentase dari missing values yang dimiliki oleh setiap kolom pada data frame
```

```python
#Menghapus (drop) setiap kolom yang mengandung missing values berdasarkan threshold
treshold = len(df) * 0.9 #Menentukan ambang batasnya
df = df.dropna(thresh=treshold, axis='columns') #menghapus berdasarkan columnya dengan acuan treshold 
df.head()
```
